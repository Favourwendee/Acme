```python
# pivot=pd.pivot_table(data=df, index=["year"], columns=['gen'], values='thousands', agg_func='sum', fill_value=0)
pivot = df.pivot_table(index=["year"], columns=['gen'], values='thousands', aggfunc='sum', fill_value=0)
```

```python
pivot
```

```python
pivot = pivot.fillna(0)
```

```python
pivot
```

```python
ax = pivot.plot.bar(stacked=True, color=["blue", "red",'green',"purple", "tomato", "lightseagreen", 'pink'], figsize=(15,6));
ax.set_title("ACME Tablet Sales (2011-2022)", fontsize=20);
ax.set_ylim(0,4000);
legend=plt.legend(title="Generation")
ax.set_xticklabels(["","2012","","2014","","2016","","2018","","2020","","2022"], rotation=0);
plt.ylabel("Unit Sold (thousands)");
plt.xlabel("Year");
```
